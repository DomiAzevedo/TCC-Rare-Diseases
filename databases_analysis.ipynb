```python
# General Information
df_info = pd.read_excel("Orphanet/en_product1.xlsx")
df_info = df_info[["id", "OrphaCode", "ExpertLink", "Name", "Synonym", "Name9",  "Name12", 
            "Source", "Reference", "Contents", "Info"]]

df_info
```

```python
import numpy as np

df = df_info.drop(['Reference', 'Source'], axis=1)

# Agrupando e agregando os sinônimos
df_aggregated = (
    df.dropna(subset=['Synonym'])  # Remove linhas com NaN em 'Sinonimo'
      .groupby('Name', as_index=False)
      .agg({'Synonym': lambda x: list(x)})
)

# Adicionando palavras sem sinônimos
all_words = pd.DataFrame({'Name': df['Name'].unique()})
df_aggregated = all_words.merge(df_aggregated, on='Name', how='left').fillna({'Synonym': np.nan})
```

```python
df_aggregated
```

```python
df_pivot = (
    df_info.pivot_table(
        index='Name',            # Agrupa pela palavra
        columns='Source',           # Usa Source como novas colunas
        values='Reference',         # Valores a serem preenchidos nas colunas
        aggfunc=lambda x: '; '.join(x.dropna()) if not x.isna().all() else None  # Combina valores ou retorna None
    )
    .reset_index()                  # Reseta o índice para ter uma tabela normal
)

# Garantindo que todas as palavras apareçam
all_words = pd.DataFrame({'Name': df_info['Name'].unique()})
df_pivot = all_words.merge(df_pivot, on='Name', how='left').fillna(np.nan)

df_pivot
```

```python
df_merged = pd.merge(df_aggregated, df_pivot, on='Name', how='outer')
df_merged.head()
```

```python
df_first = (
    df_info.groupby('Name', as_index=False)
    .first()  # Pega a primeira ocorrência para cada palavra
)

# Garantindo todas as palavras únicas
all_words = pd.DataFrame({'Name': pd.concat([df_info['Name'], df_aggregated['Name']]).unique()})
df_pivot = all_words.merge(df_pivot, on='Name', how='left')

# Juntando tudo: códigos, sinônimos e colunas extras
df_merged = pd.merge(df_aggregated, df_pivot, on='Name', how='outer')
df_merged = pd.merge(df_first[['OrphaCode', 'Name', 'ExpertLink', 'Name9', 'Name12', 'Contents', 'Info']], df_merged, on='Name', how='left')
```

```python
df_merged = df_merged.rename(columns={'Name9': 'Category', 'Name12': 'ClassificationLevel', 'Synonym': 'Synonyms'})
df_merged.head()
```

```python
df_merged.to_csv('data_info.csv', index=False)
```
